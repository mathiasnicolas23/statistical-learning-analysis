Por cuadrados minimos

```python
def pearson_chi_sq(y, f):
  # https://en.wikipedia.org/wiki/Pearson%27s_chi-squared_test
  return np.sum((y-f)**2/f)

def reduced_chi_sq(y, f, sd, m):
  # https://en.wikipedia.org/wiki/Reduced_chi-squared_statistic
  return np.sum(((y-f)/sd)**2) / (len(y)-m)

def r_sq(y, f):
  # https://en.wikipedia.org/wiki/Coefficient_of_determination
  rss = np.sum((y-f)**2)
  tss = np.sum((y-np.mean(y))**2)
  return 1 - rss/tss
  # the following not always holds true
  # ess = np.sum((f-np.mean(y))**2)
  # return ess/tss
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x_tiempo = np.array([0, 0.825, 1.275, 1.565,1.755, 1.88,2])
y_alturamax = np.array([1.2, 0.37, 0.15, 0.06, 0.03, 0.01,0])
x_tiempofrio = np.array([0, 0.76, 1.07, 1.225, 1.31,2])
y_alturamaxfrio = np.array([1.2, 0.22, 0.05, 0.01, 0,0])
```

```python
plt.figure(figsize=(8,5),dpi=200)
plt.scatter(x_tiempo,y_alturamax)
plt.scatter(x_tiempofrio,y_alturamaxfrio)
plt.plot(x_tiempo, y_alturamax, label = 'Altura máxima con temperatura ambiente')
plt.plot(x_tiempofrio,y_alturamaxfrio, label = 'Altura máxima con temperatura baja')
plt.title('Gráfico de alturas máximas')
plt.xlabel('Tiempo(s)')
plt.ylabel('Altura máxima(m)')

plt.legend()
plt.show()
```

```python
x_hn_menos1 = np.array([0.37, 0.15, 0.06,0.03])
y_hn = np.array([ 0.15, 0.06,0.03, 0.01])
x_80 = np.array([ 0.28, 0.12, 0.05, 0.02])
y_80 = np.array([0.12, 0.05, 0.02, 0])
x_40 = np.array([ 0.17, 0.07,0.03, 0.01])
y_40 = np.array([ 0.07,0.03, 0.01, 0.01])
```

```python
def calculaHN(hnm1,r):
  return (r**2)*hnm1

popt, pcov = curve_fit(calculaHN, x_hn_menos1, y_hn )

popt80, pcov80 = curve_fit(calculaHN, x_80, y_80 )
popt40, pcov40 = curve_fit(calculaHN, x_80, y_80 )

coefR = popt[0]

dR = np.sqrt(pcov[0,0])



print('  Coeficiente R = ', coefR)
print(' dR = ', dR)
```

```python
axis = np.linspace(0, 0.4, 1000)
plt.figure(figsize=(8,5),dpi=200)
plt.scatter(x_hn_menos1,y_hn, color="red")
plt.scatter(x_80,y_80, color="orange")
plt.scatter(x_40,y_40,color="purple" )
plt.plot(axis, calculaHN(axis, *popt), label="Ajuste altura inicial 1.2 m", linewidth=1, color="red")
plt.plot(axis, calculaHN(axis, *popt80), label="Ajuste altura inicial 0.8 m", linewidth=1, color="orange")
plt.plot(axis, calculaHN(axis, *popt40), label="Ajuste altura inicial 0.4 m", linewidth=1, color="purple")


plt.title('$h_n$ en función de $h_{n-1}$')
plt.xlabel('$h_{n-1}$ (m)')
plt.ylabel('$h_{n}$ (m)')

plt.legend()
plt.show()
```
